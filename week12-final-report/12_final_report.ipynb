## 2. Publication-Quality Figures

```python
# Figure 1: Model comparison bar chart
# Replace with actual numbers
models = ['LR', 'RF', 'GBM', 'LSTM', 'CNN+LSTM', 'CNN+LSTM\n(SG)', 'CNN-TF', 'CNN-TF\n+UQ']
mae_vals = [25, 20, 18, 16, 14, 13, 12, 11]  # placeholder

fig, ax = plt.subplots(figsize=(10, 5))
colors = plt.cm.Blues(np.linspace(0.3, 0.9, len(models)))
bars = ax.bar(models, mae_vals, color=colors, edgecolor='white')

# Annotate bars
for bar, val in zip(bars, mae_vals):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            f'{val:.1f}', ha='center', va='bottom', fontsize=10)

ax.set_ylabel('MAE (cycles)')
ax.set_title('RUL Prediction — Model Comparison (FD001)')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout()
plt.savefig('../reports/model_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Figure 2: Prediction scatter plot (best model)
# Placeholder — replace with actual values
np.random.seed(42)
y_true = np.random.randint(0, 130, 100)
y_pred = y_true + np.random.randn(100) * 15

fig, ax = plt.subplots(figsize=(7, 7))
ax.scatter(y_true, y_pred, alpha=0.5, c='steelblue', s=40)
ax.plot([0, 130], [0, 130], 'k--', alpha=0.5, label='Perfect prediction')
ax.set_xlabel('True RUL (cycles)')
ax.set_ylabel('Predicted RUL (cycles)')
ax.set_title('CNN-Transformer: Predicted vs. True RUL')
ax.legend()
ax.set_xlim(0, 140)
ax.set_ylim(0, 140)
plt.tight_layout()
plt.savefig('../reports/pred_vs_true.png', dpi=300, bbox_inches='tight')
plt.show()
```
